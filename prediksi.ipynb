```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_squared_log_error, r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
M5 = pd.read_csv("M5_World_Championship.csv")
```

```python
# menampilkan beberapa baris dari sebuah dataframe
M5.head()
```

```python
# Menampilkan informasi singkat
M5.info()
```

```python
# Menemukan data yang terduplikat
M5.duplicated() .sum()
```

```python
# Menampilkan deskripsi dari dataframe
M5.describe()
```

```python
# mengidentifikasi nilai null
M5.isnull() .sum()
```

```python
# Encode kolom 'Hero' menjadi nilai numerik
label_encoder = LabelEncoder()
# Menggunakan LabelEncoder untuk kolom 'Hero'
M5['Hero_Encoded'] = label_encoder.fit_transform(M5['Hero'])
M5['Roles_Encoded'] = label_encoder.fit_transform(M5['Roles'])
```

```python
# Identifikasi nilai non-numerik dalam kolom yang menyebabkan error
problematic_columns = ['T_PickPercentage', 'BS_WinRate', 'RS_WinRate', 'T_BansPercentage', 'T_PicksBansPercentage', 'T_WinRate']

for col in problematic_columns:
    # Ganti nilai non-numerik dengan nilai NaN (Not a Number)
    M5[col] = pd.to_numeric(M5[col], errors='coerce')

# Konversi tipe data object ke float
M5['T_PickPercentage'] = M5['T_PickPercentage'].astype(float)
M5['BS_WinRate'] = M5['BS_WinRate'].astype(float)
M5['RS_WinRate'] = M5['RS_WinRate'].astype(float)
M5['T_BansPercentage'] = M5['T_BansPercentage'].astype(float)
M5['T_PicksBansPercentage'] = M5['T_PicksBansPercentage'].astype(float)
M5['T_WinRate'] = M5['T_WinRate'].astype(float)
```

```python
# Memilih kolom dengan tipe data objek
obj_columns = M5.select_dtypes(include=['object']).columns
print("Kolom dengan tipe data objek:")
print(obj_columns)
```

```python
# Hapus kolom-kolom dengan tipe data objek
M5 = M5.drop(obj_columns, axis=1)
```

```python
M5.info()
```

```python
# mengidentifikasi nilai null
M5.isnull() .sum()
```

```python
# Menghapus beberapa kolom sekaligus
columns_to_drop = ['BS_WinRate', 'RS_WinRate', 'T_WinRate', 'T_PickPercentage', 'T_BansPercentage', 'T_PicksBansPercentage']
M5 = M5.drop(columns_to_drop, axis=1)
```

```python
# Menghitung korelasi antar variabel numerik
correlation_matrix = M5.corr()
print(correlation_matrix)
```

```python
# Pilih kolom-kolom untuk dimasukkan ke dalam X dan Y
features = ['Hero_Encoded', 'T_Picked', 'BS_Picked', 'RS_Picked', 'T_Banned', 'T_PicksBans',]  # Fitur untuk X
target = 'T_Wins'  # Target untuk Y

# Pisahkan data independent (X) dan dependent (Y)
X = M5[features]  # Memilih kolom-kolom fitur
Y = M5[target]  # Memilih kolom target
```

```python
# Tampilkan data X dan Y
print("Data Independent (X):")
print(X.head())

print("\nData Dependent (Y):")
print(Y.head())
```

```python
# Memisahkan data menjadi data pelatihan (train) dan data uji (test)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=100)
```

```python
print (X_train.shape)
print (X_test.shape)
```

```python
# Inisialisasi dan pelatihan model Random Forest
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, y_train)
```

```python
# membuat prediksi menggunakan RandomForest
y_pred = rf_regressor.predict(X_test)
```

```python
# Metrik evaluasi untuk regresi
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
msle = mean_squared_log_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Squared Log Error (MSLE): {msle:.4f}")
print(f"R-squared (R2): {r2:.2f}")
```
